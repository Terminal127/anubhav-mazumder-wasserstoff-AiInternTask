```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Load a pretrained YOLOv8n model
model = YOLO("runs/segment/train/weights/best.pt")
source = input("Enter image path: ")

# Run inference on the source
results = model("TACO:-Trash-Annotations-in-Context-Dataset-16/valid/images/000000_jpg.rf.088c8e7e06a599a4369f3076e1dc1c6a.jpg")  # list of Results objects

# Display the image
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3))
img = cv2.imread(source)
ax1.imshow(img)
ax1.set_title("Original Image")
ax1.axis('off')
ax2.imshow(results[0].plot())
ax2.set_title("Detection")
ax2.axis('off')
plt.tight_layout()
plt.show()
```
